```python
import requests
from bs4 import BeautifulSoup

# Function to fetch HTML content from a URL
# This function will store the html content got the url being passed
def get_html_content(url):
    response = requests.get(url)
    return response.content

# Function to extract information from HTML using BeautifulSoup

## This code is used to extract the url of the courses for each school
def extract_titles(html_content,x1,y1):
    soup = BeautifulSoup(html_content, 'html.parser')
    titles = soup.find_all(attrs={'data-id': lambda x: x and x1 <= int(x) <= y1})  # Adjust the tag and class based on the structure of the website
    links = [element.find('a')['href'] for element in titles]

    course_links=[]
    for link in links:
        course_links.append(link)
        
    return course_links

# This function is used to extract the course links for each school
def main1(x,y):
    url = 'https://classes.usc.edu/term-20233/'  # Replace with the URL of the website you want to scrape
    html_content = get_html_content(url)
    titles = extract_titles(html_content,x,y)
    print(titles)
    return titles
    
# This function is used to extract the data of each course
def main2(url):
    html_content = get_html_content(url)
    titles = extract_courses(html_content)
    return titles

# Function to extract information from HTML using BeautifulSoup
def extract_courses(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    course_data = []
    course_elements = soup.find_all("div", class_="course-info expandable")

    for course in course_elements:
        course_id = course["id"]
        course_link = course.find("a", class_="courselink")["href"]
        course_title = course.find("a", class_="courselink").text.split(":")[1].split("(")[0]
        course_description = course.find("div", class_="catalogue").text
                
        # Find the element with text "Corequisite"
        corequisite_element = soup.find("li", class_="coreq")

        # Initialize an empty list to store co-requisites
        corequisites = []

        # Check if the "Corequisite" element exists
        if corequisite_element:
            # Extract the text within the "Corequisite" element
            corequisite_text = corequisite_element.text

            # Find all links within the "Corequisite" element
            coreq_links = corequisite_element.find_all("a")

            # Iterate over each link to extract the href and text
            for coreq_link in coreq_links:
                coreq_href = coreq_link["href"]
                coreq_text = coreq_link.text.strip()

                # Append the co-requisite information to the list
                corequisites.append((coreq_href, coreq_text))


        course_data.append({
            "course_id": course_id,
            "course_link": course_link,
            "course_title": course_title,
            "course_description": course_description,
            "corequisites":corequisites
        })

    return course_data

viterbi_schhol_course_links=main1(133,148)
# print(viterbi_schhol_course_links)

course_data1=[]
for item in viterbi_schhol_course_links:
    course_data1 += main2(item)

keck_schhol_course_links=main1(178,204)
# print(keck_schhol_course_links)

course_data2=[]
for item in keck_schhol_course_links:
    course_data2 += main2(item)
    
course_data_final = course_data1 + course_data2
print("COURSE DATA Final",course_data_final)
```

```python
import json

# Specify the file path
file_path = 'output.json'

# Write to the file
with open(file_path, 'w') as json_file:
    json.dump(course_data_final, json_file, indent=2)

print(f"JSON data has been written to {file_path}")
```
